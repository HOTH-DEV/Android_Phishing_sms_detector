# Exploration du dataset de phishing
EDA : répartition des classes, longueurs, mots fréquents.

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('../data/processed/dataset.csv')
print(df.shape)
df.head()
```

```python
df['label'].value_counts().plot.bar(title='Répartition des classes')
plt.show()
```

```python
df['len'] = df['text'].str.len()
df.groupby('label')['len'].plot.hist(alpha=0.5, legend=True, bins=40, title='Longueur du texte')
plt.show()
```

```python
df['type'].value_counts().plot.pie(autopct='%1.0f%%', title='Types d\'entrée')
plt.show()
```
